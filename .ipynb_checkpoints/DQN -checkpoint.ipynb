```python
from keras.models import Sequential
from keras.layers import *
from keras.optimizers import Adam
from keras import backend as K
import tensorflow as tf
import random
import numpy as np
import gym
from collections import deque
import os
import matplotlib.pyplot as plt
```

```python
EP = 500
REWARD_HISTORY = []
```

```python
class Config:
    img_size = (210, 160, 3)
    dropout_rate = 0.75
    #learning rate alpha
    lr = 3e-5
    action_choices = 3

    #Exploration vs Exploitation
    epsilon = 1.
    epsilon_min = 0.1
    epsilon_decay = 0.99

    #Gamma of Markov decision process
    gamma = 0.95
  
```

```python

class DQAgent:
    def __init__(self, config):
        self.config = config
        self.model = self._build_model()
        #This method would clone the architecture as well as the initial weights of the base model into target model        
        self.target_model = self._build_model()
        self.memory = deque(maxlen=2000)
        self.model.summary()

    def _huber_loss(self, target, pred, clip_delta=1.):
    
    #Huber loss is a loss function used in robust regression,
    #that is less sensitive to outliers in data than the squared error loss
    
        error = K.abs(target - pred)
        cond = error < clip_delta
        loss = tf.where(cond, 0.5 * K.square(error), clip_delta * (error - 0.5 * clip_delta))
        return K.mean(loss)

    def _build_model(self):
        '''
        Model is built in this method
        '''
        bottle_seq = [
            Conv2D(32,(5,5),padding='valid', activation='relu', input_shape=self.config.img_size),
            MaxPooling2D(),
            Conv2D(64,(3,3),padding='valid', activation='relu'),
            MaxPooling2D(),
            Conv2D(128,(3,3),padding='valid', activation='relu'),
            MaxPooling2D(),
            Flatten()
        ]

        action_seq = [
            Dense(128,activation='relu'),
            Dropout(self.config.dropout_rate),
            Dense(64, activation='relu'),
            Dropout(self.config.dropout_rate),
            Dense(self.config.action_choices, activation='linear')
        ]
        model = Sequential(bottle_seq + action_seq)
        model.compile(optimizer=Adam(self.config.lr), loss=self._huber_loss)
        return model

    def update_target_model(self):
        '''
        This method would update weights of target model
        '''
        self.target_model.set_weights(self.model.get_weights())

    def act(self, state):
        '''
        Get the next action using epsilon greedy policy for deciding whether to exploit or explore
        '''
        if random.uniform(0, 1) < self.config.epsilon:
            return random.randrange(self.config.action_choices)
        else:
            #q-values
            return np.argmax(self.model.predict(state)[0])


    def replay(self, batch_size):
        '''
        Does the back propogation to adjust weights during exploitation action. 
        '''
        if batch_size > len(self.memory):
            return
    
         # Sample random minibatch of transition from replay memory
        minibatch = random.sample(self.memory, batch_size)
        for state, action, reward, next_state, done in minibatch:
            target = self.model.predict(state)
            if done:
                target[0][action] = reward
            else:
                a = self.model.predict(next_state)[0]
                t = self.target_model.predict(next_state)[0]
                target[0][action] = reward + self.config.gamma * np.max(t)
            self.model.fit(state, target, epochs=1, verbose=0)
        if self.config.epsilon > self.config.epsilon_min:
            self.config.epsilon *= self.config.epsilon_decay

    def remember(self, state, action, reward, next_state, done):
        '''
        Store the experience in our replay memory
        '''
        self.memory.append((state, action, reward, next_state, done))

    def save_model(self, path):
        self.model.save_weights(path)

    def load_model(self, path):
        self.model.load_weights(path)
```

```python
 def preprocess(frame):
    '''
    This method resizes the frame to 84*84 and converts it to grayscale
    '''
    resized = cv2.resize(frame, (210, 160), interpolation = cv2.INTER_AREA)
    frame_gray = rgb2gray(resized)
    return frame_gray
```

```python
#Name of environment to be used
env = gym.make('Breakout-v0')
state_space = env.observation_space.shape
print("state_size:",env.observation_space.shape)
action_size = env.action_space.n
print("Action size:",action_size)
done = False
batch_size = 32
rew_list = []
```

```python
print("state_size:",state_space)
print("Action size:",action_size)
print("q table size:",state_space*action_size)
```

```python
config = Config()
config.action_choices = action_size
config.img_size = state_space
agent = DQAgent(config)

# done = False
# batch_size = 32
# rew_list = [] 


for e in range(EP):
    state = env.reset()
    state = state.reshape((1,) + state_space)
    point = 0
    # # processed_frame = preprocess(state)
    # frames = []
    # rgb_frames = []
    # rgb_frames.append(state)
    for t in range(500):
        action = agent.act(state)
      
        next_state, reward, done, _ = env.step(action)  
        # if(e % 25 == 0):
        #     rgb_frames.append(next_state)
        #reward  = reward if not done else -10
        #REWARD CLIPPING IS DONE
        if(reward > 0):
            reward = 1
        elif(reward < 0):
            reward = -1
        point += reward
        next_state = next_state.reshape((1,) + state_space)
        agent.remember(state, action, reward, next_state, done)
        state = next_state
        if done:
            REWARD_HISTORY.append(point)
            agent.update_target_model()
            rew_list.append(point)
            if(e % 25 == 0):
                print("Finished episode ", e , "/", EP, " Total reward = ", sum(rew_list)/len(rew_list))
          
            print('episode {}/{}, score: {}'.format(e, EP, point))
            break
    agent.replay(batch_size)
    if(e % 10 == 0):
      print("e = ", e, " eps = ", config.epsilon)
```

###3. What are the rewards? Why did you choose them? (5 Points)

A reward represents the change in game score. Note that in general the game score may depend on the
whole prior sequence of actions and observations; In my code, IThe agent will receive 500 times rewards, and since the scale of scores varies greatly from game to game, I
fixed all positive rewards to be 1 and all negative rewards to be −1, leaving 0 rewards unchanged.
Clipping the rewards in this manner limits the scale of the error derivatives and makes it easier to
use the same learning rate across multiple games.

```python
class DQAgent2(DQAgent):
  
    def replay(self, batch_size):
        '''
        Does the back propogation to adjust weights during exploitation action. 
        '''
        if batch_size > len(self.memory):
            return
    
         # Sample random minibatch of transition from replay memory
        minibatch = random.sample(self.memory, batch_size)
        for state, action, reward, next_state, done in minibatch:
            target = self.model.predict(state)
            if done:
                target[0][action] = reward
            else:
                a = self.model.predict(next_state)[0]
                t = self.target_model.predict(next_state)[0]
                target[0][action] = reward + self.config.gamma * np.median(t) #Changing the policy
            self.model.fit(state, target, epochs=1, verbose=0)
        if self.config.epsilon > self.config.epsilon_min:
            self.config.epsilon *= self.config.epsilon_decay
```

### 8. Does Q-learning use value-based or policy-based iteration? (5 Points)

Q-learning is an off-policy algorithm,in other words it uses value-based iteration. An off-policy algorithm is an algorithm that, during training, uses a behaviour policy (that is, the policy it uses to select actions) that is different than the optimal policy it tries to estimate (the optimal policy).𝑄-learning often uses an 𝜖-greedy policy to behave (that is, to exploit and explore the environment), which uses the maximum action value as

###10. What is meant by the expected lifetime value in the Bellman equation?(5 Points)Explain, not a yes or no question. 

Expected lifetime value in the Bellman equation is the value that accumulates in each episode during the running period of algorithm. In the Bellman equation, the value function $V(t)$depends on the value function $V(t+1)$. Besides, the value of $V(t)$ can be obtained before the state reaches time t+1.

###12. How does SARSA differ from Q-learning? (5 Points) Details including pseudocode and math.




In Sarsa learning, agent starts in state $s_t$, execute action at,then get reward rt, so agent transfers to the state $s_{t+1}$, before
updating Q($s_t$, t), agent have carried out another action $a_{t+1}$
and got reward $r_{t+1}$. In contrast, in Q-learning, agent starts
in state $s_t$, execute action $a_t$, then get reward $r_t$, then check
in state $s_{t+1}$ and take action a, which  a = ${argmax}_a$Q($s_{t+1}$, a),
then update Q($s_t$, $a_t$). So the difference between this two
algorithm is the way the future rewards are found. In the Q-learning, the future rewards is the maximum expected value
of which agent is most likely to take the most advantageous
action, whereas in SARSA, it is the actual reward for executing real action.The update rules are as follows:

###14. Explain the SARSA algorithm. (5 Points)  Details including pseudocode and math.

SARSA is an on-policy TD control method. A policy is a state-action pair tuple. The Q value for a state-action is updated by an error, adjusted by the learning rate alpha. Q values represent the possible reward received in the next time step for taking action a in state s, plus the discounted future reward received from the next state-action observation.
