```python
import numpy as np
import matplotlib.pyplot as plt
from pdb import set_trace

stationary=True
class Feedback():  #Creating a feedback environment
  def __init__(self, cus_count):
    """
    Customer' feedback with rewards 1 or 0.
    
    At initialization, customers' feedbacks are created. The probability of each feedback
    returning reward 1 if customer replied from Bernouilli(p), where p randomly
    chosen from Uniform(0,1) at initialization
    """
    self.cus_count = cus_count
    self.generate_thetas()
    self.timestep = 0
    global stationary
    self.stationary=stationary
    
  def generate_thetas(self):
    self.thetas = np.random.uniform(0,1,self.cus_count)
  
  def get_reward_regret(self, cus_index):
    """ Returns random reward for cus_index action. Assumes actions are 0-indexed
    Args:
      cus_index_name is an int
    """
    self.timestep += 1
    if (self.stationary==False) and (self.timestep%100 == 0) :
      self.generate_thetas()
    # Simulate bernouilli sampling
    sim = np.random.uniform(0,1,self.cus_count)
    rewards = (sim<self.thetas).astype(int)
    reward = rewards[cus_index]
    regret = self.thetas.max() - self.thetas[cus_index]
    
    return reward, regret
```

### Thompson sampling



*The cell below are referenced from the notebook by Andre Cianflone - Thompson sampling https://colab.research.google.com/drive/1BHVH712x2Q2As9E5nN5Y8UR74T8w6AMO*

```python
class BetaAlgo():
  """
  The algos try to learn which customer is the best to maximize reward.
  
  It does this by modelling the distribution of the feedback cus_indexs with a Beta, 
  assuming the true probability of success of an cus_index is Bernouilli distributed.
  """
  def __init__(self, feedback):
    """
    Args:
      feedback: the feedback class the algo is trying to model
    """
    self.feedback = feedback
    self.cus_count = feedback.cus_count
    self.alpha = np.ones(self.cus_count)
    self.beta = np.ones(self.cus_count)
  
  def get_reward_regret(self, cus_index):
    reward, regret = self.feedback.get_reward_regret(cus_index)
    self._update_params(cus_index, reward)
    return reward, regret
  
  def _update_params(self, cus_index, reward):
    self.alpha[cus_index] += reward
    self.beta[cus_index] += 1 - reward  
    "0/1"

class BernGreedy(BetaAlgo):
  def __init__(self, feedback):
    super().__init__(feedback)
  
  @staticmethod
  def name():
    return 'beta-greedy'
   
  def get_action(self):
    """ Bernouilli parameters are the expected values of the beta"""
    theta = self.alpha / (self.alpha + self.beta)
    return theta.argmax()
  
class BernThompson(BetaAlgo):
  def __init__(self, feedback):
    super().__init__(feedback)

  @staticmethod
  def name():
    return 'thompson a=1'
  
  def get_action(self):
    """ Bernouilli parameters are sampled from the beta"""
    theta = np.random.beta(self.alpha, self.beta)
    return theta.argmax()
```

### E-greedy

*The cell below are referenced from the notebook by Andre Cianflone - Thompson sampling https://colab.research.google.com/drive/1BHVH712x2Q2As9E5nN5Y8UR74T8w6AMO*

```python
epsilon = 0.1
class EpsilonGreedy():
  """
  Epsilon Greedy with incremental update.
  Based on Sutton and Barto pseudo-code, page. 24
  """
  def __init__(self, feedback):
    global epsilon
    self.epsilon = epsilon
    self.feedback = feedback
    self.cus_count = feedback.cus_count
    self.Q = np.zeros(self.cus_count) # q-value of actions
    self.N = np.zeros(self.cus_count) # action count
  
  @staticmethod
  def name():
    return 'epsilon-greedy'
  
  def get_action(self):
    if np.random.uniform(0,1) > self.epsilon:
      action = self.Q.argmax()
    else:
      action = np.random.randint(0, self.cus_count)
    return action
  
  def get_reward_regret(self, cus_index):
    reward, regret = self.feedback.get_reward_regret(cus_index)
    self._update_params(cus_index, reward)
    return reward, regret
  
  def _update_params(self, cus_index, reward):
    self.N[cus_index] += 1 # increment action count
    self.Q[cus_index] += 1/self.N[cus_index] * (reward - self.Q[cus_index]) # inc. update rule
 
  
```

### **UCB**

*The cell below are referenced from the notebook by Andre Cianflone - Thompson sampling https://colab.research.google.com/drive/1BHVH712x2Q2As9E5nN5Y8UR74T8w6AMO*

```python
ucb_c = 2
class UCB():
  """
  Epsilon Greedy with incremental update.
  Based on Sutton and Barto pseudo-code, page. 24
  """
  def __init__(self, feedback):
    global ucb_c
    self.ucb_c = ucb_c
    self.feedback = feedback
    self.cus_count = feedback.cus_count
    self.Q = np.zeros(self.cus_count) # q-value of actions
    self.N = np.zeros(self.cus_count) + 0.0001 # action count
    self.timestep = 1
  
  @staticmethod
  def name():
    return 'ucb'
  
  def get_action(self):
    ln_timestep = np.log(np.full(self.cus_count, self.timestep))
    confidence = self.ucb_c * np.sqrt(ln_timestep/self.N)
    action = np.argmax(self.Q + confidence)
    self.timestep += 1
    return action
  
  def get_reward_regret(self, cus_index):
    reward, regret = self.feedback.get_reward_regret(cus_index)
    self._update_params(cus_index, reward)
    return reward, regret
  
  def _update_params(self, cus_index, reward):
    self.N[cus_index] += 1 # increment action count
    self.Q[cus_index] += 1/self.N[cus_index] * (reward - self.Q[cus_index]) # inc. update rule
```

### Random sampling



This simply selects a variant at random for each trial, it does not  learn, it does not update. This is purely to benchmark the other agents.

```python
'''Definition: Random sampling is a part of the sampling technique in which each sample has an equal probability of being chosen. 
A sample chosen randomly is meant to be an unbiased representation of the total population. '''

class RandomSampler():
    def __init__(self, feedback):
      self.feedback = feedback
      self.cus_count = feedback.cus_count
      self.Q = np.zeros(self.cus_count) # q-value of actions
      self.N = np.zeros(self.cus_count)

    def name ():
      return 'Random Sampling'


    def get_reward_regret(self, cus_index):
        reward, regret = self.feedback.get_reward_regret(cus_index)
        self._update_params(cus_index, reward)
        return reward, regret


    def get_action(self):
        # ln_timestep = np.log(np.full(self.cus_count, self.timestep))
        # confidence = self.ucb_c * np.sqrt(ln_timestep/self.N)
        # action = np.argmax(self.Q + confidence)
        # self.timestep += 1
        action = np.random.choice(self.cus_count)
        return action

    def _update_params(self, cus_index, reward):
        self.N[cus_index] += 1 # increment action count
        self.Q[cus_index] += reward # inc. update rule
```

##**3 >Applying Models**





*The cell below are referenced from the notebook by Andre Cianflone - Thompson sampling https://colab.research.google.com/drive/1BHVH712x2Q2As9E5nN5Y8UR74T8w6AMO*

```python

def plot_data(y):
  """ y is a 1D vector """
  x = np.arange(y.size)
  _ = plt.plot(x, y, 'o')
  
def multi_plot_data(data, names):
  """ data, names are lists of vectors """
  x = np.arange(data[0].size)
  for i, y in enumerate(data):
    plt.plot(x, y, 'o', markersize=2, label=names[i])
  plt.legend(loc='upper right', prop={'size': 12}, numpoints=10)
  plt.show()
  
def simulate(simulations, timesteps, cus_count, Algorithm):
  """ Simulates the algorithm over 'simulations' epochs """
  sum_regrets = np.zeros(timesteps)
  for e in range(simulations):
    feedback = Feedback(cus_count)
    algo = Algorithm(feedback)
    regrets = np.zeros(timesteps)
    for i in range(timesteps):
      action = algo.get_action()
      reward, regret = algo.get_reward_regret(action)
      regrets[i] = regret
    sum_regrets += regrets  
  mean_regrets = sum_regrets / simulations
  return mean_regrets

def experiment(cus_count, timesteps=1000, simulations=1000):
  """ 
  Standard setup across all experiments 
  Args:
    timesteps: (int) how many steps for the algo to learn the feedback
    simulations: (int) number of epochs
  """
  algos = [EpsilonGreedy, UCB, BernThompson, RandomSampler]
  regrets = []
  names = []
  for algo in algos:
    regrets.append(simulate(simulations, timesteps, cus_count, algo))
    names.append(algo.name())
  multi_plot_data(regrets, names)
```

```python

# Experiment 1
cus_count = 10 # number of customer in feedback
epsilon = 0.1
ucb_c = 2
stationary=True
experiment(cus_count)
```

Since there are so many actions, and since the mean rewards are evenly distributed between 0 and 1, there is a high cost of constantly exploring actions with rewards close to 0. Noticeably, the Thompson algorithm is incredibly stable across time, despite the large action space. **Exploration and Exploitation**

The algorithm is a method of real-time decision-making, striving to strike a balance between fully exploring the variable space and using the best operation.
It is important to strike a balance between the two. First, you need to fully explore the variant space in order to determine the strongest variant. By first determining and then continuing to utilize the best action, you will maximize the total rewards available from the environment. However, you also want to continue to explore other viable variants in case they provide better returns in the future. In other words, you want to hedge your bets by continuing to (slightly) experiment with sub-optimal variants when the payout changes. If they do, your algorithm will accept the change and start choosing this variant for new customers. Another benefit of exploring is that you can learn more about the generation process of variants. In other words, what is its average payment rate and what is the uncertainty distribution around it. Therefore, the key is to decide how to best balance this exploration-exploitation trade-off.

###5. How long do Thompson Sampling, e-greedy, ucb, and random sampling remember the past actions? Explain your answer. (10 Points) 

**"Thompson Sampling"**


Thompson sampling remrembers the past actions by implements the idea of probability matching. Because its reward estimations Q̃  are sampled from posterior distributions, each of these probabilities is equivalent to the probability that the corresponding action is optimal, conditioned on observed history.

An action is chosen by first sampling from the beta distribution, followed by choosing the action with highest mean reward:

𝑥𝑡=argmax𝑘(𝜃̂ 𝑘),𝜃̂ 𝑘∼beta(𝛼𝑘,𝛽𝑘)

According to Bayes' rule, an action's posterior distribution is updated depending on the reward 𝑟𝑡 received:

(𝛼𝑘,𝛽𝑘)=(𝛼𝑘,𝛽𝑘)+(𝑟𝑡,1−𝑟𝑡)

Thus the actions' posterior distribution are constantly updated throughout the simulation. 

**"e-greedy"**

The ε-greedy algorithm takes the best action most of the time, but does random exploration occasionally. The action value is estimated according to the past experience by averaging the rewards associated with the target action a that we have observed so far (up to the current time step t):
For example :
to update the $Q$-value of each action, we maintain a count of each action. For action $k$ taken at time $t$:

$$
\begin{align}
r_t &\leftarrow \text{observe}(k) \\
N(k) &\leftarrow N(k) + 1 \\
Q(k) &\leftarrow Q(k) + \frac{1}{N(k)}[r_t-Q(k)] \\
\end{align}
$$

**"ucb"**

The Upper Confidence Bounds (UCB) algorithm measures this potential by an upper confidence bound of the reward value.

UCB favors exploration of actions with a strong potential to have a optimal value.

During the processing, the model will automatically remermber the prevopis rewords.The selected action follows the rule:

$$
A_t = \text{argmax}_a \left[ Q_t(a) + c \sqrt{\frac{\ln t}{N_t (a)}} \right]
$$

a = action selected (in our case the feedback given)

Nt(a) = number of times action a was selected prior to the time t

Qt(a) = average reward of action a prior to the time t

Both Nt(a) and Qt(a) will help remember the previous steps as well as rewards






**"Random Sampling"**

Since random sampling is a part of the sampling technique in which each sample has an equal probability of being chosen, thus I think this algorithm will not remember the actions before , nor learn from the past actions. It just add each time reward.
